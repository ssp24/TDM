```python
import re
```

```python
document_text = open ('testfile.txt', 'r') 

text_string = document_text.read()
text_string1 = text_string.lower()
```

```python
lines = [] #Erstellt eine leere Liste
for line in text_string1.splitlines():
    lines.append(line) #befüllt die Liste mit der jeweiligen Zeile
    
```

```python
#Läuft für jede Zeile durch und gibt den Ort des Treffers an (aber nicht den Inhalt)

for element in lines:
    a = re.match('file', element)
    print(a)
```

```python
for element in lines:
    a = re.match('file', element)
    if a: 
        print(a)
      
```
