```python
import requests
import pandas as pd
import lxml.etree as ET
from bs4 import BeautifulSoup as soup
import xml.etree.ElementTree as ET
import unicodedata


df = pd.read_csv("alldnb_bilingual_ger.csv")
df.head()
```

```python
#Spalte "identifier" als eigenes Dataframe: 

df_id = pd.DataFrame(df, columns=['identifier'])
df_id
```

```python
#Funktion zur Übergabe des Inhalttextes: 

def retrieve_text(lst, text):
    try:
        return [i for i in lst if text in i][0]
    except:
        return None

    
```

```python
#Ergänzen Spalten mit Teilen des Textes aus df_id anhand der String-Funktion: 

df_id['ISBN_test'] = df_id['identifier'].str.split().apply(lambda x: retrieve_text(x, 'ISBN')) 
df_id['IDN_test'] = df_id['identifier'].str.split().apply(lambda x: retrieve_text(x, 'IDN'))
```

```python
df_id
```

```python
#Einfügen neuer Spalten "ISBN" und "IDN", in dem die Spalten "ISBN_test" und "IDN_test" übernommen, aber in deren 
#Inhaltstext die Buchstaben "ISBN:" bzw. "IDN:" entfernt werden: 

df_id['ISBN'] = df_id['ISBN_test'].str.replace('ISBN:', '')
df_id['IDN'] = df_id['IDN_test'].str.replace('IDN:', '')
df_id
```

```python
#Entfernen der Spalten "identifier", "ISBN_test" und "IDN_test":

df_id = df_id.drop(columns=['identifier', 'ISBN_test', 'IDN_test'])
df_id
```

Da das ursprüngliche Dataframe bereits eine Spalte "ISBN" enthält, vergleichen wir nun die Angaben der bereits vorhandenen Spalte mit den von uns aus der Spalte "identifier" gefilterten Informationen zur ISBN: 

```python
df['ISBN'].isin(df_id['ISBN']).value_counts()
```

Wir sehen, dass es in einigen Fällen Unterschiede gibt und die Spalten leider nicht immer identisch sind. Wir entscheiden uns daher hier dafür, die ursprünglich enthaltenen Spalte "ISBN" beizubehalten, jedoch eine neue Spalte, die wir "ISBN_new" nennen werden, hinzuzufügen, die den ISBN-Text aus dem Feld "identifier" enthalten wird: 

```python
#Ergänze Spalten "ISBN" und "IDN" im ursprünglichen Dataframe: 

df.insert(loc=3, column='ISBN_new', value=df_id['ISBN'])
df.insert(loc=4, column='IDN', value=df_id['IDN'])
df
```

```python
df.to_csv("alldnb_bilingual_ger_id.csv")
```

Falls uns weiterhin interessiert, an welchen Stellen der beiden Dataframes für die ISBNs die Unterschiede bestehen, können wir die beiden Dataframes anhand der Spalten "ISBN" zusammenführen und uns im Anschluss die Einträge anzeigen lassen, an denen die ISBNs nicht übereinstimmten: 

```python

df_diff  = pd.merge(df, df_id, on="ISBN", how='outer', indicator='Exist')
df_diff  = df_diff.loc[df_diff['Exist'] != 'both']
df_diff
```
