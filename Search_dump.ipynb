```python
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
import unicodedata
```

```python
#load file with BeuatifulSoup: 

with open('data/dataset_tutorial_eco.xml', encoding='utf-8') as f:
    contents = f.read()
    xml = BeautifulSoup(contents, 'lxml')
```

```python
records = xml.find_all('record', {'type':'Bibliographic'})
```

```python
#print(xml.text[0:100])
#print(contents[0:1000])
print(len(records))
#print(records[1])
```

```python
#Function für Titeldaten in MARC21
def get_stuff(item):

    ns = {"marc":"http://www.loc.gov/MARC21/slim"}
    xml = etree.fromstring(unicodedata.normalize("NFC", str(item)))
    
    #idn
    idn = xml.findall("controlfield[@tag = '001']", namespaces=ns)
    if idn:
        idn = idn[0].text
    else:
        idn = 'N/A' 
        
    
    #idn
    creator = xml.findall("datafield[@tag = '100']/subfield[@code = 'a']", namespaces=ns)
    if creator:
        creator = creator[0].text
    else:
        creator = 'N/A' 
        
        
    #link
    link = xml.findall("datafield[@tag = '856']/subfield[@code = 'u']")
    link2 = xml.findall("datafield[@tag = '856']/subfield[@code = 'u']")[1].text
    if link:
        link = link[0].text
    else:
        link = 'N/A' 

   

    
    meta_dict = {"IDN":idn, "creator":creator, "link":link, "link2":link2}
    
    return meta_dict
```

```python
result = [get_stuff(record) for record in records]
df = pd.DataFrame(result)
df
```
